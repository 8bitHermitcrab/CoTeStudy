stack = 스택

nums = 숫자리스트(오름차순)

i = nums에 쓸 index

result = 목표 수열

cnt = result에 쓸 index

​

1. stack이 비어있으면 값을 추가한다

2. 비어있지 않으면 스택 마지막값을 result[cnt](목표수열 cnt번째와 비교합니다)

3. 같으면 stack을 pop 후 '-' 저장합니다

4. 다르면 i가 answer 값과 같은지 비교합니다(스택에 값을 전부 넣었는지 판단)

5. i값과 answer값이 같으면 더 이상 진행이 불가하므로 NO를 출력하고 종료합니다

6. 다르면 스택에 값을 추가합니다

만들 수열은 [ 4, 3, 6 , 8, 7, 5, 2, 1] 이라 가정

1,2,3,4 push → stack = [1,2,3,4]

4,3 pop → stack = [1,2]

5,6 push → stack = [1,2,5,6]

6 pop → stack = [1,2,5]

....

수열과 pop한 순서가 일치하면 종료

```python
#https://blog.naver.com/hanjo1515/222499722591


n = int(input())

stack = []
result = []
nums = [x+1 for x in range(n)]  #아마도 인덱스를 이용하기 위해서
answer = []

#입력
for _ in range(n):
    result.append(int(input()))


i = 0
cnt = 0
while(True):
    print('==================')
    print(f'i = {i}')

    #빈 stack에 추가하기
    
    #stack 비어있으면 추가 -> push
    if stack == []:
        stack.append(nums[i])
        answer.append('+')
        i += 1  #인덱스 하나씩 증가하면서
    else:
        #여기서부터 이해가 잘 안되요ㅜ
        #pop
        if stack[len(stack)-1] == result[cnt]:
            stack.pop()
            answer.append('-')
            cnt += 1
        else:
            if i == n:
                print('NO')
                break
            stack.append(nums[i])
            answer.append('+')
            i += 1
    
    #print(f'cnt = {cnt}')
    print(f'stack = {stack}')
    print(f'answer = {answer}')

    #출력
    if cnt == n:
        print(*answer, sep = '\n')
        break
```
